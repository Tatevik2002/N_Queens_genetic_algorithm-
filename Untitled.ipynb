```python
import numpy as np
```

```python
class genetic_algorithm:
    def __init__(self, n,m):
        self.n = n
        self.m = m
        
    def initialize(self):
        initial_population =[ np.zeros((self.n,self.n)) for i in range(self.m) ]
        for i in range(self.m): #m individuals
            if i == self.m-1:
                t = 0
                d = 0
                while np.count_nonzero(initial_population[i])<self.n:
                    initial_population[i][0][t]= 1.
                    if self.n%2==1 and np.count_nonzero(initial_population[i]) == self.n:
                        continue 
                    initial_population[i][self.n-1-d][self.n-1]= 1.
                    t+=1
                    d+=1


            while np.count_nonzero(initial_population[i])<self.n:
                l = np.random.randint(0,self.n)
                j = np.random.randint(0,self.n)
                initial_population[i][l][j] = 1.

        return initial_population
    
    def fitness(self,chess_board): #chess board is a matrix  of zeroes and ones
        # returns how many  queens are under attack 
        coordinates_ = np.argwhere(chess_board)

        unique, counts_x = np.unique(coordinates_[:,0:1],return_counts=True)

        rep_x = np.where(counts_x>1,counts_x,0)


        repeating_x_quantity= np.sum(rep_x*(rep_x-1)/2) # all possible collisions

        unique,counts_y = np.unique(coordinates_[:,1:2],return_counts=True)

        rep_y = np.where(counts_y>1,counts_y,0)

        repeating_y_quantity= np.sum(rep_y*(rep_y-1)/2) # all possible collisions

        diag =  0

        for i in range(0,len(chess_board)-1):
            a = np.diagonal(chess_board,i)
            unique, counts = np.unique(a, return_counts=True)
            if (len(counts)>1):
                if counts[1]>1: # 1.0 is the second element of the counts  
                    diag = diag + counts[1]*(counts[1]-1)/2
            else:
                if a[0] == 1.0:
                    diag = diag+ len(a)*(len(a)-1)/2 # all a are queens

        for i in range(1,len(chess_board)-1):
            a = np.diagonal(chess_board,-i)
            unique, counts = np.unique(a, return_counts=True)
            if (len(counts)>1):
                if counts[1]>1:
                    diag = diag + counts[1]*(counts[1]-1)/2
            else:
                if a[0] == 1.0:
                    diag = diag+ len(a)*(len(a)-1)/2 # all a are queens

        flipped_board = np.fliplr(chess_board)
        for i in range(0,len(flipped_board)-1):
            a = np.diagonal(flipped_board,-i)
            unique, counts = np.unique(a, return_counts=True)
            if (len(counts)>1):
                if counts[1]>1:
                    diag = diag + counts[1]*(counts[1]-1)/2
            else:
                if a[0] == 1.0:
                    diag = diag+ len(a)*(len(a)-1)/2 # all a are queens
        for i in range(1,len(flipped_board)-1):
            a = np.diagonal(flipped_board,-i)
            unique, counts = np.unique(a, return_counts=True)
            if (len(counts)>1):
                if counts[1]>1:
                    diag = diag + counts[1]*(counts[1]-1)/2
            else:
                if a[0] == 1.0:
                    diag = diag+ len(a)*(len(a)-1)/2 # all a are queens

        # it is not possible that both x and y are the same for two queens
        return -(repeating_x_quantity+repeating_y_quantity+diag)-1 # minus one because  we don't want fittness score to be zero

    def selection(self,population): # population is a 3D matrix containing boards
    # returns one parent
        fitness_scores = np.array([])
        for i in range(len(population)):
            np.append(fitness_scores,fitness(population[i]))
        sum_ = np.sum(fitness_scores)
        probabilities = fitness_scores/sum_
        parent_index = np.random.choice(np.arange(len(population)),probabilities)
        return np.array(population[parent_index])
    
    def crossover(self,parent1,parent2): # single point crossover
        coordinates_1 = np.argwhere(parent1)
        coordinates_2 = np.argwhere(parent2)
        random_point = np.random.randint(0,len(coordinates_1)-1)
        offspring1_coordinate = np.concatenate((coordinates_1[:random_point+1],coordinates_2[random_point+1:]),axis  =0)
        offspring2_coordinate = np.concatenate((coordinates_2[:random_point+1],coordinates_1[random_point+1:]),axis  =0)
        offspring1 = np.zeros(parent1.shape)
        offspring2 = np.zeros(parent2.shape)
        offspring1[offspring1_coordinate[:,0:1],offspring1_coordinate[:,1:2]] = np.ones(offspring1[offspring1_coordinate[:,0:1],offspring1_coordinate[:,1:2]].shape)
        offspring2[offspring1_coordinate[:,0:1],offspring1_coordinate[:,1:2]] = np.ones(offspring1[offspring1_coordinate[:,0:1],offspring1_coordinate[:,1:2]].shape)
        n = len(offspring1)
        while np.count_nonzero(offspring1)<n:
            offspring1[np.random.randint(0,len(coordinates_1)-1)][np.random.randint(0,len(coordinates_1)-1)] = 1.
        while np.count_nonzero(offspring2)<n:
            offspring2[np.random.randint(0,len(coordinates_1)-1)][np.random.randint(0,len(coordinates_1)-1)] = 1.


        return offspring1, offspring2
    
    def mutation(self,offspring): # swap mutation
        coordinates = np.argwhere(offspring)
        ind = np.random.randint(0,self.n)
        offspring[coordinates[ind][0],coordinates[ind][1]] = 0.
        while np.count_nonzero(offspring)<len(offspring):
            i = np.random.randint(0,self.n)
            j = np.random.randint(0,self.n)
            if i != coordinates[ind][0] and j != coordinates[ind][1]:
                offspring[i][j] = 1.
    #     index1x= np.random.randint(0,len(offspring))
    #     index1y = np.random.randint(0,len(offspring))
    #     index2x= np.random.randint(0,len(offspring))
    #     index2y = np.random.randint(0,len(offspring))
    #     temp = offspring[index2x,index2y]
    #     offspring[index2x,index2y] = offspring[index1x,index1y]
    #     offspring[index1x,index1y] = temp
        return offspring
    
    def max_fitness(self,population):
        scores = np.array([fitness(population[i]) for i in range(self.m)])
        return np.ndarray.max(scores)
    
    def loop(self):
        population  = self.initialize()
        q = 0
        while max_fitness(population)<-2 : 
            parent1 = self.selection(population)
            parent2 = self.selection(population)
            offspring1, offspring2 = self.crossover(parent1,parent2)
            threshold = np.random.uniform()
            if threshold >0.1:
                offspring1 = self.mutation(offspring1)

            threshold = np.random.uniform()
            if threshold >0.1:
                offspring1 = self.mutation(offspring2)
            fitness_scores = np.array([])
            k,j = 0,0
            min_k, min_j = 0,0
            for i in range(len(population)):
                score = self.fitness(population[i])
                np.append(fitness_scores,score) 
                if min_k>score:
                    k = i
                    min_k = score
                else:
                    if min_j>score:
                        j = i
                        min_j = score


            population[k] = offspring1
            population[j] = offspring2
            print(q)
            q+=1
        return population











```
